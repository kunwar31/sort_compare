```python
from time import time,sleep
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from sorting import mergeSort,bubbleSort,selectionSort,insertionSort,quickSort
```

```python
bubble_sort = {}
selection_sort = {}
insertion_sort = {}
merge_sort = {}
quick_sort = {}
python_sort = {}

TEST_RANGE = 100000
STEP_SIZE = 500

print('Testing Bubble sort')
sleep(1)
for i in tqdm(range(0,TEST_RANGE//12,STEP_SIZE)): # Takes Forever
    a = np.random.randint(low=0,high=100,size=i)
    t = time()
    bubbleSort(a)    
    bubble_sort[i] = time()-t

print('Testing Selection sort')
sleep(1)
for i in tqdm(range(0,TEST_RANGE//9,STEP_SIZE)):
    a = np.random.randint(low=0,high=100,size=i)
    t = time()
    selectionSort(a)    
    selection_sort[i] = time()-t
    
print('Testing Insertion sort')
sleep(1)
for i in tqdm(range(0,TEST_RANGE//9,STEP_SIZE)):
    a = np.random.randint(low=0,high=100,size=i)
    t = time()
    insertionSort(a)    
    insertion_sort[i] = time()-t
    
print('Testing Merge sort')
sleep(1)
for i in tqdm(range(0,TEST_RANGE,STEP_SIZE)):
    a = np.random.randint(low=0,high=100,size=i)
    t = time()
    mergeSort(a,0,len(a)-1)    
    merge_sort[i] = time()-t
    
print('Testing Quick sort')
sleep(1)
for i in tqdm(range(0,TEST_RANGE//3,STEP_SIZE)):
    a = np.random.randint(low=0,high=100,size=i)
    t = time()
    quickSort(a,0,len(a)-1)    
    quick_sort[i] = time()-t
    
print('Testing Python sort')
sleep(1)
for i in tqdm(range(0,TEST_RANGE,STEP_SIZE)):
    a = np.random.randint(low=0,high=100,size=i)
    t = time()
    a.sort()    
    python_sort[i] = time()-t
```

```python
bubble_x, bubble_y = zip(*bubble_sort.items())
selection_x, selection_y = zip(*selection_sort.items()) 
insertion_x, insertion_y = zip(*insertion_sort.items()) 
merge_x, merge_y = zip(*merge_sort.items()) 
quick_x, quick_y = zip(*quick_sort.items()) 
python_x, python_y = zip(*python_sort.items()) 
```

```python
plt.figure(figsize=(16, 12))
plt.plot(bubble_x, bubble_y)
plt.plot(selection_x, selection_y)
plt.plot(insertion_x, insertion_y)
plt.plot(merge_x, merge_y)
plt.plot(quick_x, quick_y)
plt.plot(python_x, python_y)
plt.legend(['Bubble Sort',
            'Selection Sort',
            'Insertion Sort',
            'Merge Sort', 
            'Quick Sort',
            'Python Sort'], loc='upper left')
plt.show()
```
